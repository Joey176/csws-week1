```python
import time
import datetime
import math
import sys
import swisseph as swe
from astral import LocationInfo
from astral.sun import sun

# --- Config & Structures ---
TATTWAS = [
    ("PRITHVI [EARTH]", 6, "[Y-SQ]"),
    ("JALA    [WATER]", 12, "[S-CR]"),
    ("TEJAS   [FIRE] ", 18, "[R-TR]"),
    ("VAYU    [AIR]  ", 24, "[B-CI]"),
    ("AKASHA  [ETHER]", 30, "[B-EG]")
]
TOTAL_TATTWA_CYCLE_MINS = 90

CHALDEAN_ORDER = ["SATURN", "JUPITER", "MARS", "SUN", "VENUS", "MERCURY", "MOON"]
WEEKDAY_LORDS = {0: "MOON", 1: "MARS", 2: "MERCURY", 3: "JUPITER", 4: "VENUS", 5: "SATURN", 6: "SUN"}
ZODIAC_SIGNS = ["ARI", "TAU", "GEM", "CAN", "LEO", "VIR", "LIB", "SCO", "SAG", "CAP", "AQU", "PIS"]

PLANET_MAP = {
    "SUN": swe.SUN, "MOON": swe.MOON, "MERCURY": swe.MERCURY, "VENUS": swe.VENUS,
    "MARS": swe.MARS, "JUPITER": swe.JUPITER, "SATURN": swe.SATURN
}

RED = "\033[1;31m"
RESET = "\033[0m"
HIDE_CURSOR = "\033[?25l"
SHOW_CURSOR = "\033[?25h"

# --- Calculation Modules ---

def get_chinese_zodiac(year):
    """Calculates the Stem-Branch elemental cycle for the Chinese Zodiac."""
    stems = ["Yang Wood", "Yin Wood", "Yang Fire", "Yin Fire", "Yang Earth", "Yin Earth", "Yang Metal", "Yin Metal", "Yang Water", "Yin Water"]
    branches = ["Rat", "Ox", "Tiger", "Rabbit", "Dragon", "Snake", "Horse", "Goat", "Monkey", "Rooster", "Dog", "Pig"]
    if year < 4: return "Unknown", "Unknown"
    stem_idx = (year - 4) % 10
    branch_idx = (year - 4) % 12
    return stems[stem_idx], branches[branch_idx]

def get_jewish_meta(now):
    """Hardcoded structural anchors for the June 2026 Lunisolar junction."""
    if now.year == 2026 and now.month == 6:
        if now.day == 15:
            return "5786.03.30", "ROSH CHODESH TAMMUZ [NEW MOON]"
    return "5786.03.15", "ORDINARY CYCLE"

def calculate_lunar_phase(julian_day):
    """Calculates moon phase illumination percentage using Swiss Ephemeris positions."""
    # Fixed the unpacking error here by assigning to two variables instead of three
    res_sun, _ = swe.calc_ut(julian_day, swe.SUN, 0)
    res_moon, _ = swe.calc_ut(julian_day, swe.MOON, 0)
    sun_long = res_sun[0]
    moon_long = res_moon[0]

    # Calculate phase angle
    diff = (moon_long - sun_long) % 360
    illumination = (1 - math.cos(math.radians(diff))) / 2 * 100

    return illumination, "SUPER NEW MOON [DARK MOON]" if illumination < 1 else "LUNAR MATRIX ACTIVE"

def get_sun_data(lat, lon):
    city = LocationInfo("Local", "Region", "UTC", lat, lon)
    now = datetime.datetime.now(datetime.timezone.utc)
    s_today = sun(city.observer, date=now.date(), tzinfo=datetime.timezone.utc)
    s_tomorrow = sun(city.observer, date=now.date() + datetime.timedelta(days=1), tzinfo=datetime.timezone.utc)
    s_yesterday = sun(city.observer, date=now.date() - datetime.timedelta(days=1), tzinfo=datetime.timezone.utc)
    return s_yesterday, s_today, s_tomorrow

def calculate_tattwa(now, sunrise_today):
    time_diff = now.astimezone(datetime.timezone.utc) - sunrise_today
    diff_minutes = time_diff.total_seconds() / 60.0
    minute_in_cycle = diff_minutes % TOTAL_TATTWA_CYCLE_MINS

    elapsed = 0
    for name, duration, symbol in TATTWAS:
        if elapsed <= minute_in_cycle < (elapsed + duration):
            return name, symbol, int((elapsed + duration) - minute_in_cycle)
        elapsed += duration
    return TATTWAS[0][0], TATTWAS[0][1], 0

def calculate_planetary_hour(now, s_yesterday, s_today, s_tomorrow):
    now_utc = now.astimezone(datetime.timezone.utc)
    if s_today['sunrise'] <= now_utc < s_today['sunset']:
        phase = "DAYTIME"; start_time = s_today['sunrise']; end_time = s_today['sunset']
        day_of_week = start_time.weekday(); base_offset = 0
    else:
        phase = "NIGHTTIME"; base_offset = 4
        if now_utc >= s_today['sunset']:
            start_time = s_today['sunset']; end_time = s_tomorrow['sunrise']; day_of_week = s_today['sunrise'].weekday()
        else:
            start_time = s_yesterday['sunset']; end_time = s_today['sunrise']; day_of_week = s_yesterday['sunrise'].weekday()

    total_duration = end_time - start_time
    hour_length = total_duration / 12
    current_hour_idx = int((now_utc - start_time) / hour_length)
    if current_hour_idx > 11: current_hour_idx = 11
    if current_hour_idx < 0: current_hour_idx = 0

    day_lord = WEEKDAY_LORDS[day_of_week]
    final_planet_idx = (CHALDEAN_ORDER.index(day_lord) + base_offset + current_hour_idx) % 7
    return current_hour_idx + 1, phase, CHALDEAN_ORDER[final_planet_idx], int(((start_time + (hour_length * (current_hour_idx + 1))) - now_utc).total_seconds() / 60), int(hour_length.total_seconds() / 60), day_lord

def get_zodiac_positions(julian_day):
    positions = {}
    for name, swe_id in PLANET_MAP.items():
        # Fixed the unpacking error here by assigning to two variables instead of three
        res, _ = swe.calc_ut(julian_day, swe_id, swe.FLG_SPEED)
        sign_index = int(res[0] // 30)
        # Checked direction marker safety indices
        retro_marker = "<-" if res[3] < 0 else "  "
        positions[name] = f"{int(res[0] % 30):02d}DG {ZODIAC_SIGNS[sign_index]} {retro_marker}"
    return positions

# --- Main Engine Loop ---

def run_alien_clock(lat, lon):
    is_colab = 'google.colab' in sys.modules
    if not is_colab:
        print(HIDE_CURSOR, end="")

    blink_state = True
    try:
        while True:
            s_yesterday, s_today, s_tomorrow = get_sun_data(lat, lon)
            now_local = datetime.datetime.now().astimezone()
            now_utc = datetime.datetime.now(datetime.timezone.utc)
            julian_day = swe.julday(now_utc.year, now_utc.month, now_utc.day, now_utc.hour + now_utc.minute/60.0 + now_utc.second/3600.0)

            # Run calculations
            t_name, t_symbol, t_rem = calculate_tattwa(now_local, s_today['sunrise'])
            p_hour, p_phase, p_ruler, p_rem, p_len, p_day_ruler = calculate_planetary_hour(now_local, s_yesterday, s_today, s_tomorrow)
            planet_positions = get_zodiac_positions(julian_day)

            # Run macro meta frameworks
            c_stem, c_branch = get_chinese_zodiac(now_local.year)
            j_chrono, j_holiday = get_jewish_meta(now_local)
            m_illum, m_phase_name = calculate_lunar_phase(julian_day)

            sep = ":" if blink_state else " "
            blink_state = not blink_state
            time_str = now_local.strftime(f"%H{sep}%M{sep}%S")

            if not is_colab:
                print("\033[H\033[J", end="")

            print(RED + "==================================================")
            print(f" ::: METAPHYSICAL CORE // TIME-MATRIX ENGINE ::: ")
            print("==================================================")
            print(f" [LOCAL SYSTEM TIME] ------> [ {time_str} ]")
            print(f" [MATRIX CHRONO] -----------> [ {now_local.strftime('%Y.%m.%d')} ]")
            print(f" [CHINESE VECTOR] ----------> [ Year of the {c_stem} {c_branch} ]")
            print(f" [HEBREW LUNISOLAR] --------> [ AM {j_chrono} // {j_holiday} ]")
            print("--------------------------------------------------")
            print(f" >> ACTIVE TATTWA ELEMENT VECTOR")
            print(f"    MATRIX IDENT: {t_name}")
            print(f"    GEOM SIGNIFIER: {t_symbol}")
            print(f"    TERM WINDOW : {t_rem:02d} MIN REMAINING")
            print("--------------------------------------------------")
            print(f" >> CURRENT PLANETARY CYCLE HORIZON")
            print(f"    INTERVAL    : HOUR {p_hour:02d} / 12 [{p_phase}]")
            print(f"    CYCLE LRD   : {p_day_ruler} [OVERALL DAY]")
            print(f"    CHRONO LRD  : {p_ruler} [ACTIVE HOUR]")
            print(f"    SYS DECAY   : {p_rem:02d} MIN TO NEXT SHIFT (SCALE: {p_len}M)")
            print("--------------------------------------------------")
            print(f" >> AXIS PROXIMITY EVENT TRACKER")
            print(f"    NEXT SOLSTICE : 2026.06.21 [SUMMER]")
            print(f"    NEXT EQUINOX  : 2026.09.22 [AUTUMNAL AXIS DISTANT]")
            print(f"    LUNAR PHASING : {m_phase_name} ({m_illum:.2f}% LUMEN)")
            print("--------------------------------------------------")
            print(f" >> CELESTIAL LONGITUDE COORDINATES")
            for planet, position in planet_positions.items():
                pointer = ">> " if planet == p_ruler else "   "
                print(f"    {pointer}{planet:<8} :: [ {position} ]")
            print("==================================================")

            if is_colab:
                print(" [SNAPSHOT RENDER COMPLETE IN CLOUD ENGINE]" + RESET)
                break

            print(" [CTRL+C TO SUSPEND INTERFACE CORE PROCESS]" + RESET)
            time.sleep(1)

    except KeyboardInterrupt:
        print(SHOW_CURSOR + RESET + "\n[INTERFACE SYSTEM OFFLINE]")
    finally:
        swe.close()

if __name__ == "__main__":
    LATITUDE = 53.4084; LONGITUDE = -2.9916
    run_alien_clock(LATITUDE, LONGITUDE)
```

```python
import time
import datetime
import math
import sys
from zoneinfo import ZoneInfo
from IPython.display import clear_output
import swisseph as swe
from astral import LocationInfo
from astral.sun import sun

# --- Config & Hierarchies ---
TATTWAS = [
    ("PRITHVI [EARTH]", 6, "[Y-SQ]"),
    ("JALA    [WATER]", 12, "[S-CR]"),
    ("TEJAS   [FIRE] ", 18, "[R-TR]"),
    ("VAYU    [AIR]  ", 24, "[B-CI]"),
    ("AKASHA  [ETHER]", 30, "[B-EG]")
]
TOTAL_TATTWA_CYCLE_MINS = 90

CHALDEAN_ORDER = ["SATURN", "JUPITER", "MARS", "SUN", "VENUS", "MERCURY", "MOON"]
WEEKDAY_LORDS = {0: "MOON", 1: "MARS", 2: "MERCURY", 3: "JUPITER", 4: "VENUS", 5: "SATURN", 6: "SUN"}
ZODIAC_SIGNS = ["ARI", "TAU", "GEM", "CAN", "LEO", "VIR", "LIB", "SCO", "SAG", "CAP", "AQU", "PIS"]

# Traditional Sefer Yetzirah & Kabbalistic Mapping
KABBALAH_MATRIX = {
    "SATURN":  {"hebrew": "SHABTAI (שבתאי)", "archangel": "CASSIEL",  "angel": "ZAPHKIEL"},
    "JUPITER": {"hebrew": "TZEDEK (צדק)",   "archangel": "SACHIEL",  "angel": "ZADKIEL"},
    "MARS":    {"hebrew": "MA'ADIM (מאדים)", "archangel": "SAMAEL",   "angel": "CAMAEL"},
    "SUN":     {"hebrew": "CHAMAH (חמה)",   "archangel": "MICHAEL",  "angel": "RAPHAEL"},
    "VENUS":   {"hebrew": "NOGAH (נוגה)",   "archangel": "ANAEL",    "angel": "HANIEL"},
    "MERCURY": {"hebrew": "KOKHAV (כוכב)",  "archangel": "RAPHAEL",  "angel": "MICHAEL"},
    "MOON":    {"hebrew": "LEVANAH (לבנה)",  "archangel": "GABRIEL",  "angel": "GABRIEL"}
}

PLANET_MAP = {
    "SUN": swe.SUN, "MOON": swe.MOON, "MERCURY": swe.MERCURY, "VENUS": swe.VENUS,
    "MARS": swe.MARS, "JUPITER": swe.JUPITER, "SATURN": swe.SATURN
}

RED = "\033[1;31m"
RESET = "\033[0m"

# --- Calculation Modules ---

def get_chinese_zodiac(year):
    stems = ["Yang Wood", "Yin Wood", "Yang Fire", "Yin Fire", "Yang Earth", "Yin Earth", "Yang Metal", "Yin Metal", "Yang Water", "Yin Water"]
    branches = ["Rat", "Ox", "Tiger", "Rabbit", "Dragon", "Snake", "Horse", "Goat", "Monkey", "Rooster", "Dog", "Pig"]
    if year < 4: return "Unknown", "Unknown"
    stem_idx = (year - 4) % 10
    branch_idx = (year - 4) % 12
    return stems[stem_idx], branches[branch_idx]

def get_jewish_meta(now):
    if now.year == 2026 and now.month == 6:
        if now.day == 15:
            return "5786.03.30", "ROSH CHODESH TAMMUZ [NEW MOON]"
    return "5786.04.02", "ORDINARY CYCLE"

def calculate_lunar_phase(julian_day):
    res_sun, _ = swe.calc_ut(julian_day, swe.SUN, 0)
    res_moon, _ = swe.calc_ut(julian_day, swe.MOON, 0)
    diff = (res_moon - res_sun) % 360
    illumination = (1 - math.cos(math.radians(diff))) / 2 * 100
    return illumination, "WAKING LUNAR ENGINE" if illumination > 1 else "DARK MOON HORIZON"

def get_sun_data(lat, lon):
    city = LocationInfo("Liverpool", "UK", "Europe/London", lat, lon)
    now = datetime.datetime.now(ZoneInfo("Europe/London"))
    s_today = sun(city.observer, date=now.date(), tzinfo=ZoneInfo("Europe/London"))
    s_tomorrow = sun(city.observer, date=now.date() + datetime.timedelta(days=1), tzinfo=ZoneInfo("Europe/London"))
    s_yesterday = sun(city.observer, date=now.date() - datetime.timedelta(days=1), tzinfo=ZoneInfo("Europe/London"))
    return s_yesterday, s_today, s_tomorrow

def calculate_tattwa(now, sunrise_today):
    time_diff = now - sunrise_today
    diff_minutes = time_diff.total_seconds() / 60.0
    minute_in_cycle = diff_minutes % TOTAL_TATTWA_CYCLE_MINS

    elapsed = 0
    for name, duration, symbol in TATTWAS:
        if elapsed <= minute_in_cycle < (elapsed + duration):
            return name, symbol, int((elapsed + duration) - minute_in_cycle)
        elapsed += duration
    return "AKASHA  [ETHER]", "[B-EG]", 0

def calculate_planetary_hour(now, s_yesterday, s_today, s_tomorrow):
    if s_today['sunrise'] <= now < s_today['sunset']:
        phase = "DAYTIME"; start_time = s_today['sunrise']; end_time = s_today['sunset']
        day_of_week = start_time.weekday(); base_offset = 0
    else:
        phase = "NIGHTTIME"; base_offset = 4
        if now >= s_today['sunset']:
            start_time = s_today['sunset']; end_time = s_tomorrow['sunrise']; day_of_week = s_today['sunrise'].weekday()
        else:
            start_time = s_yesterday['sunset']; end_time = s_today['sunrise']; day_of_week = s_yesterday['sunrise'].weekday()

    total_duration = end_time - start_time
    hour_length = total_duration / 12
    current_hour_idx = int((now - start_time) / hour_length)
    if current_hour_idx > 11: current_hour_idx = 11
    if current_hour_idx < 0: current_hour_idx = 0

    day_lord = WEEKDAY_LORDS[day_of_week]
    final_planet_idx = (CHALDEAN_ORDER.index(day_lord) + base_offset + current_hour_idx) % 7
    return current_hour_idx + 1, phase, CHALDEAN_ORDER[final_planet_idx], int(((start_time + (hour_length * (current_hour_idx + 1))) - now).total_seconds() / 60), int(hour_length.total_seconds() / 60), day_lord

def get_zodiac_positions(julian_day):
    positions = {}
    for name, swe_id in PLANET_MAP.items():
        res, _ = swe.calc_ut(julian_day, swe_id, swe.FLG_SPEED)
        longitude = res
        speed = res
        sign_index = int(longitude // 30)
        retro_marker = "<-" if speed < 0 else "  "
        heb_label = KABBALAH_MATRIX[name]["hebrew"]
        positions[name] = f"{int(longitude % 30):02d}DG {ZODIAC_SIGNS[sign_index]} {retro_marker} | {heb_label}"
    return positions

# --- Main Engine Loop ---

def run_alien_clock(lat, lon):
    is_colab = 'google.colab' in sys.modules
    blink_state = True

    try:
        while True:
            # Time synchronization
            uk_tz = ZoneInfo("Europe/London")
            now_local = datetime.datetime.now(uk_tz)
            now_utc = datetime.datetime.now(datetime.timezone.utc)
            julian_day = swe.julday(now_utc.year, now_utc.month, now_utc.day, now_utc.hour + now_utc.minute/60.0 + now_utc.second/3600.0)

            # System matrix executions
            s_yesterday, s_today, s_tomorrow = get_sun_data(lat, lon)
            t_name, t_symbol, t_rem = calculate_tattwa(now_local, s_today['sunrise'])
            p_hour, p_phase, p_ruler, p_rem, p_len, p_day_ruler = calculate_planetary_hour(now_local, s_yesterday, s_today, s_tomorrow)
            planet_positions = get_zodiac_positions(julian_day)

            c_stem, c_branch = get_chinese_zodiac(now_local.year)
            j_chrono, j_holiday = get_jewish_meta(now_local)
            m_illum, m_phase_name = calculate_lunar_phase(julian_day)

            # Extracting Kabbalistic Metadata Safely
            day_meta = KABBALAH_MATRIX[p_day_ruler]
            hour_meta = KABBALAH_MATRIX[p_ruler]

            sep = ":" if blink_state else " "
            blink_state = not blink_state
            time_str = now_local.strftime(f"%H{sep}%M{sep}%S")

            # Reset visual rendering pane
            if is_colab:
                clear_output(wait=True)
            else:
                print("\033[H\033[J", end="")

            # Force output cache buffer to flush immediately to the browser
            sys.stdout.flush()

            print(RED + "==================================================")
            print(f" ::: METAPHYSICAL CORE // TIME-MATRIX ENGINE ::: ")
            print("==================================================")
            print(f" [LOCAL SYSTEM TIME] ------> [ {time_str} ] (BST)")
            print(f" [MATRIX CHRONO] -----------> [ {now_local.strftime('%Y.%m.%d')} ]")
            print(f" [CHINESE VECTOR] ----------> [ Year of the {c_stem} {c_branch} ]")
            print(f" [HEBREW LUNISOLAR] --------> [ AM {j_chrono} // {j_holiday} ]")
            print("--------------------------------------------------")
            print(f" >> ACTIVE TATTWA ELEMENT VECTOR")
            print(f"    MATRIX IDENT: {t_name}")
            print(f"    GEOM SIGNIFIER: {t_symbol}")
            print(f"    TERM WINDOW : {t_rem:02d} MIN REMAINING")
            print("--------------------------------------------------")
            print(f" >> CURRENT PLANETARY CYCLE HORIZON")
            print(f"    INTERVAL    : HOUR {p_hour:02d} / 12 [{p_phase}]")
            print(f"    CYCLE LRD   : {day_meta['hebrew']} [OVERALL DAY]")
            print(f"    SOVEREIGN   : ARCHANGEL {day_meta['archangel']} [DAY RULER]")
            print(f"    CHRONO LRD  : {hour_meta['hebrew']} [ACTIVE HOUR]")
            print(f"    CHRONO WATCH: ANGEL {hour_meta['angel']} [HOUR RULER]")
            print(f"    SYS DECAY   : {p_rem:02d} MIN TO NEXT SHIFT (SCALE: {p_len}M)")
            print("--------------------------------------------------")
            print(f" >> AXIS PROXIMITY EVENT TRACKER")
            print(f"    NEXT SOLSTICE : 2026.06.21 [SUMMER]")
            print(f"    NEXT EQUINOX  : 2026.09.22 [AUTUMNAL AXIS DISTANT]")
            print(f"    LUNAR PHASING : {m_phase_name} ({m_illum:.2f}% LUMEN)")
            print("--------------------------------------------------")
            print(f" >> SEFER YETZIRAH LONGITUDE COORDINATES")
            for planet, position in planet_positions.items():
                pointer = ">> " if planet == p_ruler else "   "
                print(f"    {pointer}{planet:<8} :: [ {position} ]")
            print("==================================================")
            print(" [CLICK STOP BUTTON TO HALT ENGINE PROCESS]" + RESET)

            time.sleep(1)

    except KeyboardInterrupt:
        print("\n[INTERFACE SYSTEM OFFLINE]")
    finally:
        swe.close()

if __name__ == "__main__":
    LATITUDE = 53.4084; LONGITUDE = -2.9916
    run_alien_clock(LATITUDE, LONGITUDE)
```
